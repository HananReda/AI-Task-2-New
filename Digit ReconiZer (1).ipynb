```python
#import libraries
import numpy as np                                          # linear algebra
import pandas as pd                                         # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt, matplotlib.image as mpimg  #Graph
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_read = pd.read_csv('../input/digitrecognizer/train.csv')   #training data 

image=train_read .iloc[0:5000,1:]                                #training data images
label=train_read .iloc[0:5000,:1]                                #training data labels
train_images,test_image,train_label,test_label=train_test_split(image,label,train_size=0.8,random_state=0)

i=1
img_draw=train_images.iloc[i].to_numpy()
img_draw=img_draw.reshape((28,28))                        #image dimensions
plt.imshow(img_draw,cmap='gray')                           #display the image
plt.title(train_label.iloc[i,0])


i=1
img_draw=train_images.iloc[i].to_numpy()
img_draw=img_draw.reshape((28,28))                        #image dimensions
plt.imshow(img_draw,cmap='gray')                         #display the image
plt.title(train_label.iloc[i,0])

#plot frequency:numbers in lables
plt.hist(train_images.iloc[i])


# training our module and find what is its efficency

clf=svm.SVC()
clf.fit(train_images,train_label.values.ravel())
clf.score(test_image,test_label)

test_read=pd.read_csv('../input/digitrecognizer/test.csv')    #testing data 

results=clf.predict(test_read[0:1000])

df = pd.DataFrame(results)
df.index.name='ImageId'
df.index+=1
df.columns=['Label']
df.to_csv('results.csv', header=True)

```
